```python
def func(str_):
    some_list=[]
    for char in str_:
        try:
            int(char)
        except:
            continue
        some_list.append(char)
    return int(''.join(some_list))

func('$100,000,00')
```

```python
def package_and_ct(raw_text):
    some_list = []
    lines = []
    
    #make a sublist for every day
    for line in raw_text:

        lines.append(line)    
        if line.split(',')[0] in week_days:
            some_list.append(lines)
            lines = []
    
    #move the date up one    
    for i in range(1, len(some_list)):
        sub = some_list[i - 1]
        next_sub = some_list[i]
        next_sub.insert(0, sub[-1])
        sub.pop()
    
    
    return some_list

raw_text = [line for line in open('/Users/ajay/Desktop/garysguide_last_year.txt')]
package_and_ct(raw_text)
```

```python
#take out the empty sublist
def clean_up(input_nested_lists):
    output = [sublist for sublist in input_nested_lists if sublist != []]
    throw_out_nothing = lambda sublist: [item.strip() for item in sublist if item.strip() != '']
    output = [throw_out_nothing(sublist) for sublist in output if sublist != []]
    return output

clean_up(package_and_ct(raw_text))
```

```python
def list_to_df(text_file_list):
    # you need to subract the row for the date out of it 
    df = pd.DataFrame([(sub_list[0], len(sub_list) - 1) for sub_list in text_file_list])
    return df

list_to_df(clean_up(package_and_ct(raw_text)))
```

```python
def filter_for_this(input_nested_lists, str_):
    filter_condition = lambda t_str, filter_str: filter_str.lower() in t_str.lower() or t_str.upper().split(',')[0] in week_days
    output = [[item for item in sublist if filter_condition(item, str_)]   \
    for sublist in input_nested_lists]
    
    return output

filter_for_this(clean_up(package_and_ct(raw_text)), 'Data Scientist')
```
